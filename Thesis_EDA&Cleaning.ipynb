```python
def outliers(dt, ft):
    Q1 = dt[ft].quantile(0.25)
    Q3 = dt[ft].quantile(0.75)
    IQR = Q3 - Q1

    LB = Q1 - 1.5 * IQR
    UB = Q3 + 1.5 * IQR

    ls = dt.index[(dt[ft] <= LB) | (dt[ft] >= UB)]
    return ls
```

```python
index_list = []
for feature in num_col:
    index_list.extend(outliers(df, feature))
```

```python
def remove(dt, ls):
    ls = sorted(set(ls))
    dt = dt.drop(index = ls)
    return dt
```

```python
df_cleaned = remove(df, index_list)
```

```python
def classify_binary(cgpa):
    if cgpa >= 2.5:
        return 1 # Passing
    else:  
        return 0 # At-Risk
        
last_df["curr_cGPA_binary"] = last_df["curr_cGPA"].apply(classify_binary)


print(last_df["curr_cGPA_binary"].value_counts())
```

```python
def classify_ternary(cgpa):
    if cgpa >= 3.25:
        return 2  # Excellent Standing
    elif 2.5 <= cgpa < 3.25:
        return 1  # Good Standing
    else:
        return 0  # At Risk

last_df["curr_cGPA_ternary"] = last_df["curr_cGPA"].apply(classify_ternary)

print(last_df["curr_cGPA_ternary"].value_counts())
```

```python
def classify_quaternary(cgpa):
    if cgpa >= 3.5:
        return 3  # Excellent Standing
    elif 3.0 <= cgpa < 3.5:
        return 2  # Good Standing
    elif 2.5 <= cgpa < 3.0:
        return 1  # Passing
    else:
        return 0  # At-Risk

last_df["curr_cGPA_quaternary"] = last_df["curr_cGPA"].apply(classify_quaternary)

print(last_df["curr_cGPA_quaternary"].value_counts())
```

```python
plt.figure(figsize=(7, 15))

plt.subplot(3, 1, 1)
sns.countplot(x=last_df["curr_cGPA_binary"], palette="cividis")
plt.title("Binary Classification - Class Distribution")
plt.xlabel("Class 0: At-Risk, 1: Passing")
plt.ylabel("Count")
for i in range(last_df["curr_cGPA_binary"].nunique()):
    count = last_df["curr_cGPA_binary"].value_counts()[i]
    plt.text(i, count + 5, str(count), ha="center")

plt.subplot(3, 1, 2)
sns.countplot(x=last_df["curr_cGPA_ternary"], palette="viridis")
plt.title("Ternary Classification - Class Distribution")
plt.xlabel("Class 0: At-Risk, 1: Good Standing, 2: Excellent Standing")
plt.ylabel("Count")
for i in range(last_df["curr_cGPA_ternary"].nunique()):
    count = last_df["curr_cGPA_ternary"].value_counts()[i]
    plt.text(i, count + 5, str(count), ha="center")

# Check class distribution for quaternary classification
plt.subplot(3, 1, 3)
sns.countplot(x=last_df["curr_cGPA_quaternary"], palette="magma")
plt.title("Quaternary Classification - Class Distribution")
plt.xlabel("Class 0: At-Risk, 1: Passing, 2: Good Standing, 3: Excellent Standing")
plt.ylabel("Count")
for i in range(last_df["curr_cGPA_quaternary"].nunique()):
    count = last_df["curr_cGPA_quaternary"].value_counts()[i]
    plt.text(i, count + 5, str(count), ha="center")

plt.show()
```
